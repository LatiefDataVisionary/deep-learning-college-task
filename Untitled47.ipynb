## Ambil ulasan

### Subtask:
Tulis kode untuk mengambil ulasan aplikasi dari Google Play Store menggunakan ID paket aplikasi.


**Reasoning**:
Write the code to scrape app reviews using the `reviews` function with the specified app ID and count, and store the result in a variable.



```python
app_id = 'com.whatsapp'
scraped_reviews, _ = reviews(
    app_id,
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort='newest', # defaults to 'relevance'
    count=100 # defaults to 100
)
```

**Reasoning**:
The traceback indicates that the `sort` parameter in the `reviews` function is expecting an object with a `value` attribute, not a string. I need to import the `Sort` enum from `google_play_scraper` and use its members for the `sort` parameter.



```python
from google_play_scraper import Sort

app_id = 'com.whatsapp'
scraped_reviews, _ = reviews(
    app_id,
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # Use Sort.NEWEST instead of 'newest'
    count=100 # defaults to 100
)
```

**Reasoning**:
The previous attempt to scrape reviews failed due to an incorrect value for the `sort` parameter. The error has been addressed by importing and using the `Sort` enum. Now, execute the corrected code to scrape the reviews.



```python
app_id = 'com.whatsapp'
scraped_reviews, _ = reviews(
    app_id,
    lang='id', # defaults to 'en'
    country='id', # defaults to 'us'
    sort=Sort.NEWEST, # Use Sort.NEWEST instead of 'newest'
    count=10000 # defaults to 100
)
```

## Tampilkan data

### Subtask:

**Reasoning**:

## Konversi data ke CSV

### Subtask:
Konversi data ulasan yang diambil menjadi DataFrame pandas dan simpan ke file CSV.

**Reasoning**:
Convert the list of dictionaries containing the scraped reviews into a pandas DataFrame for easier manipulation and then save it as a CSV file.

```python
import pandas as pd

reviews_df = pd.DataFrame(scraped_reviews)

# Save the DataFrame to a CSV file
reviews_df.to_csv('whatsapp_reviews.csv', index=False)

print("Data ulasan telah berhasil disimpan ke dalam file 'whatsapp_reviews.csv'.")
```
